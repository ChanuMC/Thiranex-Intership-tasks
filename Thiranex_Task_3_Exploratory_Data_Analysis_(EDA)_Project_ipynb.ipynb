```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/content/cleaned_job_dataset.csv")
print(df.head())
```

```python
print(df.describe())
```

```python
plt.figure(figsize=(10,6))

sns.countplot(
    y='job_category',
    data=df,
    order=df['job_category'].value_counts().index
)

plt.title("Job Category Distribution")

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.barplot(
    x='experience_level',
    y='salary_in_usd',
    data=df
)

plt.title("Average Salary by Experience Level")

plt.show()
```

```python
df['work_setting'].value_counts().plot(
    kind='pie',
    autopct='%1.1f%%'
)

plt.title("Work Setting Distribution")

plt.ylabel("")

plt.show()
```

```python
sns.countplot(
    x='company_size',
    data=df
)

plt.title("Company Size Distribution")

plt.show()
```

```python
salary_trend = df.groupby(
    'work_year'
)['salary_in_usd'].mean()

salary_trend.plot(marker='o')

plt.title("Salary Trend Over Years")

plt.xlabel("Year")

plt.ylabel("Average Salary")

plt.show()
```

```python
plt.figure(figsize=(8,6))

sns.heatmap(
    df.corr(numeric_only=True),
    annot=True,
    cmap='coolwarm'
)

plt.title("Correlation Heatmap")

plt.show()
```

```python
sns.boxplot(
    x=df['salary_in_usd']
)

plt.title("Salary Outliers")

plt.show()
```
